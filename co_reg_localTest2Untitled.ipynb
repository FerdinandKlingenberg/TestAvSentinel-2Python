```python
# visualize geometric quality grid with INITIAL shifts present in your input target image
%matplotlib inline

CRL.view_CoRegPoints(figsize=(15,15),backgroundIm='ref', savefigPath='F:/c/arosicsTest/result/',savefigDPI=300)
```

```python
# visualize geometric quality grid with shifts present AFTER shift correction
CRL_after_corr = COREG_LOCAL(im_reference, CRL.path_out, **kwargs, CPUs=1)
CRL_after_corr.view_CoRegPoints(figsize=(15,15),backgroundIm='ref', savefigPath='F:/c/arosicsTest/resultAfterCoReg/',savefigDPI=300)
```

```python
CRL_after_corr.correct_shifts()
```
